# Calculate degree of synchronisation

```python
def fraction_initiated_origins_per_cascade(initiation_times, n_ori_init, row, duration):
    cascade_event_list = []
    new_cluster = 1
    n_ori = 1
    for indx in range(0, initiation_times.size):
#         print(indx)
        if new_cluster == 1:
            count = 1
            t_max = initiation_times[indx] + duration
            n_ori = n_ori_init[indx]
            new_cluster = 0
        else:
            if initiation_times[indx] <= t_max:
                count = count + 1
            else:
#                 print('fraction of origins initiated:', count/ n_ori)
                cascade_event_list.append(count/ n_ori)
                count = 1
                t_max = initiation_times[indx] + duration
                n_ori = n_ori_init[indx]
#     print(row.sub_simu_name, cascade_event_list)   
#     print('init times:', np.array(initiation_times))
    return cascade_event_list

def mean_and_var_of_fraction_initiated_origins_per_cascade(row, data_frame_traces, duration):
    data_frame_init = pd.read_hdf(row.path_dataset, key='dataset_init_events')
    initiation_times = data_frame_init['t_init']
    n_ori_init = data_frame_init['n_ori_init']
    fract_init_origins_per_cascade = fraction_initiated_origins_per_cascade(initiation_times, n_ori_init,  row, duration)
    return np.mean(fract_init_origins_per_cascade), np.var(fract_init_origins_per_cascade)

def degree_synch(row, duration):
    data_frame_traces = pd.read_hdf(row.path_dataset, key='dataset_init_events')
    if not data_frame_traces.abort_simulation[0]:
        degree_synch, variance_synch = mean_and_var_of_fraction_initiated_origins_per_cascade(row, data_frame_traces, duration)
        return degree_synch
    else:
        return np.NaN

def variance_synch(row):
    data_frame_traces = pd.read_hdf(row.path_dataset, key='dataset_init_events')
    if not data_frame_traces.abort_simulation[0]:
        degree_synch, variance_synch = mean_and_var_of_fraction_initiated_origins_per_cascade(row, data_frame_traces)
        return variance_synch
    else:
        return 0
```

```python
def generate_dict(degree_synch, t_onset_RIDA, cascade_duration):
    return {
        "degree_synch": degree_synch,
        "licensing_time": t_onset_RIDA,
        "cascade_duration": cascade_duration}

def generate_data_frame(data_frame_, cascade_durations_):
    dictionary_list = []
    for cascade_indx in cascade_durations_:
        for df_index, df_row in data_frame_.iterrows():
            degree_synch_row = degree_synch(df_row, cascade_indx)
            dictionary = generate_dict(degree_synch_row, df_row["t_onset_RIDA"], cascade_indx)
            dictionary_list.append(dictionary)
    return dictionary_list
        
def make_legend_hill_coef_n(n):
    return r'n={}'.format(int(n))

cascade_durations = np.arange(0.1, 1, 0.1) * doubling_time
print(cascade_durations)

data_frame_small = pd.DataFrame(generate_data_frame(data_frame, cascade_durations))
```

# Make Legend

```python
def make_legend(cascade_duration, doubling_time):
    return r'$\tau_{{i}}={} \, \tau_{{d}}$'.format(np.round(cascade_duration/doubling_time, 2))
    
data_frame_small.loc[:, 'legend_cascade_duration'] = data_frame_small.apply(lambda row: make_legend(row.cascade_duration, doubling_time), axis = 1)
```
