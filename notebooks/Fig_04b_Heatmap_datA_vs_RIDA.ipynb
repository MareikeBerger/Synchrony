# Fraction of initiated origins

```python
doubling_time = 1/data_frame["doubling_rate"][0]
print(doubling_time)

cascade_duration = 0.4*doubling_time
```

```python
def frac_init_oris_per_cascade(initiation_times, n_ori_init, cascade_duration, row):
    cascade_event_list = []
    new_cluster = 1
    n_ori = 1
    for indx in range(0, initiation_times.size):
        if new_cluster == 1:
            count = 1
            t_max = initiation_times[indx] + cascade_duration
            n_ori = n_ori_init[indx]
            new_cluster = 0
        else:
            if initiation_times[indx] <= t_max:
                count = count + 1
            else:
                cascade_event_list.append(count/ n_ori)
                count = 1
                t_max = initiation_times[indx] +  cascade_duration
                n_ori = n_ori_init[indx]
    return cascade_event_list

def mean_and_var_of_frac_init_oris_per_cascade(row, data_frame_traces, cascade_duration):
    data_frame_init = pd.read_hdf(row.path_dataset, key='dataset_init_events')
    initiation_times = data_frame_init['t_init']
    n_ori_init = data_frame_init['n_ori_init']
    fract_init_origins_per_cascade = frac_init_oris_per_cascade(initiation_times, 
                                                                n_ori_init, 
                                                                cascade_duration, 
                                                                row)
    return np.mean(fract_init_origins_per_cascade), np.var(fract_init_origins_per_cascade)

def degree_synch(row, cascade_duration):
    data_frame_traces = pd.read_hdf(row.path_dataset, key='dataset_init_events')
    if not data_frame_traces.abort_simulation[0]:
        degree_synch, variance_synch = mean_and_var_of_frac_init_oris_per_cascade(row, 
                                                                                  data_frame_traces, 
                                                                                  cascade_duration)
        return degree_synch
    else:
        return np.NaN

def variance_synch(row, cascade_duration):
    data_frame_traces = pd.read_hdf(row.path_dataset, key='dataset_init_events')
    if not data_frame_traces.abort_simulation[0]:
        degree_synch, variance_synch = mean_and_var_of_frac_init_oris_per_cascade(row, 
                                                                                  data_frame_traces, 
                                                                                  cascade_duration)
        return variance_synch
    else:
        return 0

data_frame.loc[:, 'degree_synch'] = data_frame.apply(lambda row: degree_synch(row, cascade_duration), axis = 1)
data_frame.loc[:, 'variance_synch'] = data_frame.apply(lambda row: variance_synch(row, cascade_duration), axis = 1)
```

# Plot degree of synchrony (new definition)

```python
def cap_degree_synch(row):
    if row.degree_synch > 3:
        return 3
    else:
        return row.degree_synch

data_frame.loc[:, 'degree_synch_capped'] = data_frame.apply(lambda row: cap_degree_synch(row), axis = 1)
```
