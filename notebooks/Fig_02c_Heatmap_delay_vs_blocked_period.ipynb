# Fraction of initiated origins

```python
def fraction_initiated_origins_per_cascade(initiation_times, n_ori_init, row):
    cascade_event_list = []
    new_cluster = 1
    n_ori = 1
    for indx in range(0, initiation_times.size):
#         print(indx)
        if new_cluster == 1:
            count = 1
            t_max = initiation_times[indx] + row.t_doubling_datA + 10 * row.time_step
            n_ori = n_ori_init[indx]
            new_cluster = 0
        else:
            if initiation_times[indx] <= t_max:
                count = count + 1
            else:
#                 print('fraction of origins initiated:', count/ n_ori)
                cascade_event_list.append(count/ n_ori)
                count = 1
                t_max = initiation_times[indx] + row.t_doubling_datA + 50 * row.time_step
                n_ori = n_ori_init[indx]
#     print(row.sub_simu_name, cascade_event_list)   
#     print('init times:', np.array(initiation_times))
    return cascade_event_list

def mean_and_var_of_fraction_initiated_origins_per_cascade(row, data_frame_traces):
    data_frame_init = pd.read_hdf(row.path_dataset, key='dataset_init_events')
    initiation_times = data_frame_init['t_init']
    n_ori_init = data_frame_init['n_ori_init']
    fract_init_origins_per_cascade = fraction_initiated_origins_per_cascade(initiation_times, n_ori_init,  row)
    return np.mean(fract_init_origins_per_cascade), np.var(fract_init_origins_per_cascade)

def degree_synch(row):
    data_frame_traces = pd.read_hdf(row.path_dataset, key='dataset_init_events')
    if not data_frame_traces.abort_simulation[0]:
        degree_synch, variance_synch = mean_and_var_of_fraction_initiated_origins_per_cascade(row, data_frame_traces)
        return degree_synch
    else:
        return np.NaN

def variance_synch(row):
    data_frame_traces = pd.read_hdf(row.path_dataset, key='dataset_init_events')
    if not data_frame_traces.abort_simulation[0]:
        degree_synch, variance_synch = mean_and_var_of_fraction_initiated_origins_per_cascade(row, data_frame_traces)
        return variance_synch
    else:
        return 0


data_frame.loc[:, 'degree_synch'] = data_frame.apply(lambda row: degree_synch(row), axis = 1)
data_frame.loc[:, 'variance_synch'] = data_frame.apply(lambda row: variance_synch(row), axis = 1)
data_frame['degree_synch']
```

# Mark aborted simulations in dataframe

```python
def determine_whether_simulation_aborted(filepath_h5):
    data_frame_traces = pd.read_hdf(filepath_h5, key='dataset_init_events')
    return data_frame_traces["abort_simulation"][0]

data_frame.loc[:, 'aborted'] = data_frame.apply(lambda row: determine_whether_simulation_aborted(row.path_dataset), axis = 1)
# data_frame["aborted"]
```

# Plot degree of synchrony (new definition)

```python
def cap_degree_synch(row):
    if row.degree_synch > 3:
        return 3
    else:
        return row.degree_synch

data_frame.loc[:, 'degree_synch_capped'] = data_frame.apply(lambda row: cap_degree_synch(row), axis = 1)
```

```python
import matplotlib.colors as colors
import matplotlib as mpl

data_frame_ordered = data_frame.sort_values(by=["t_doubling_datA"])
heatmap_df = data_frame_ordered.pivot("t_doubling_datA", "period_blocked", "degree_synch")
print(data_frame["degree_synch"].min(), data_frame["degree_synch"].max())
heatmap_mask = data_frame_ordered.pivot("t_doubling_datA", "period_blocked", "aborted")

# divnorm = colors.TwoSlopeNorm(vmin=data_frame["degree_synch"].min(), vcenter=1, vmax=2)
cmap = mpl.cm.get_cmap("seismic").copy()
cmap.set_over('darkred')
sns.set(rc={'figure.figsize': (18.7, 3.27)})
ax = sns.heatmap(heatmap_df,
            mask=heatmap_mask,  
            square=True, 
            linewidths=.5,
            cmap=cmap,
            vmin=0, 
            vmax=2, 
            center=1.00,
            cbar_kws={'extend': 'max'})
ax.invert_yaxis()

# ax.set_title(r'$s= \frac{n_{\rm ori}(t_{\rm i})}{n_{\rm ori}(t_{\rm f})- n_{\rm ori}(t_{\rm i})}$')
ax.set_facecolor('grey')
ax.set(xlabel=r'$\tau_{\rm b}}$ [h]', ylabel=r'$\tau_{\rm l}$ [h]')
plt.savefig(file_path + '/heatmap_degree_synch_def.png', format='png', bbox_inches='tight', dpi=400)
```
